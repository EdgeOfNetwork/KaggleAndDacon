```python
import numpy as np
import pandas as pd
#from pycaret.classification import *

cloud_path = '/content/drive/MyDrive/0.competition/dataset/'

path = cloud_path

labeled = pd.read_csv(path + 'cn7_labeled.csv',index_col=0)
unlabeled = pd.read_csv(path + 'cn7_unlabeled.csv', index_col=0)
```

```python
labeled
```

```python
labeled_data = labeled.sample(frac=0.99, random_state=2021)
labeled_unseen_data = labeled.drop(labeled_data.index)

labeled_data.reset_index(drop=True, inplace=True)
labeled_unseen_data.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(labeled_data.shape))
print('Unseen Data For Predictions: ' + str(labeled_unseen_data.shape))
```

```python
from pycaret.anomaly import *

data = setup(labeled_data, normalize = True, session_id=1)
```
